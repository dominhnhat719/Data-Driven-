```python
import pandas as pd
import matplotlib.pyplot as plt

# c file CSV
file_path = "C:\\Users\\Admin\\Desktop\\Volve Field\\NO_15_9-F-1_C.csv"   # thay bng ng dn file ca bn
df = pd.read_csv(file_path)

# Chuyn DATEPRD v datetime
df['DATEPRD'] = pd.to_datetime(df['DATEPRD'], errors='coerce')

# To ct "days" = s ngy k t ngy u tin
df['days'] = (df['DATEPRD'] - df['DATEPRD'].min()).dt.days

# V ch s Downhole Pressure (v d ging hnh bn gi)
plt.figure(figsize=(16,4))
plt.plot(df['days'], df['AVG_DOWNHOLE_PRESSURE'], color="blue")

plt.xlabel("Time (days)")
plt.ylabel("Downhole Pressure")
plt.title("Volve Field - downhole pressure")

# t ticks  gia khung hnh (100, 200, , 700)
plt.xticks(ticks=range(0, 800, 100))  

plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
