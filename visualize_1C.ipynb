```python
import pandas as pd
import matplotlib.pyplot as plt

# Đọc file CSV
file_path = "C:\\Users\\Admin\\Desktop\\Volve Field\\NO_15_9-F-1_C.csv"   # thay bằng đường dẫn file của bạn
df = pd.read_csv(file_path)

# Chuyển DATEPRD về datetime
df['DATEPRD'] = pd.to_datetime(df['DATEPRD'], errors='coerce')

# Tạo cột "days" = số ngày kể từ ngày đầu tiên
df['days'] = (df['DATEPRD'] - df['DATEPRD'].min()).dt.days

# Vẽ chỉ số Downhole Pressure (ví dụ giống hình bạn gửi)
plt.figure(figsize=(16,4))
plt.plot(df['days'], df['AVG_DOWNHOLE_PRESSURE'], color="blue")

plt.xlabel("Time (days)")
plt.ylabel("Downhole Pressure")
plt.title("Volve Field - downhole pressure")

# Đặt ticks ở giữa khung hình (100, 200, …, 700)
plt.xticks(ticks=range(0, 800, 100))  

plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
